## 4.3. PyTorchプログラミングにおける用語定義

- レイヤー関数
    - テンソルを受け取り，テンソルを返す関数
    - 主に使うレイヤー関数



- パラメータ
    - レイヤー関数が内部で持っている，入力以外のデータ
    - `nn.Linear`, `nn.Conv2d`, `nn.BatchNorm2d`がパラメータを持つ
- 入力テンソル
- 出力テンソル
- 機械学習モデル
    - レイヤー関数を組み合わせて，入力テンソルを受け取り，出力テンソルを返す合成関数
- 学習
    - レイヤー関数内部のパラメータを，調整すること
    - 勾配降下法などの最適化関数が用いられる
